```python
import pandas as pd
import pickle
from sklearn.cluster import KMeans
from tqdm import tqdm
```

```python
y_cluster_ty_mth_2=pickle.load(open('__y_cluster_100_ty_mth_2.pkl','rb'))
y_cluster_ty_mth_2=y_cluster_ty_mth_2.sort_values(by='cluster_100')
y_cluster_ty_mth_2=y_cluster_ty_mth_2.reset_index(drop=True)
y_cluster_ty_mth_2
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
import time
from selenium.webdriver.common.keys import Keys
import time

no_bisocial=list(y_cluster_ty_mth_2['no_bisocial'])
```

```python
driver=webdriver.Chrome("/Users/kimminji/company_clustering_project/chromedriver")
cols=['bisocial','기업명','업종','기업규모','기업형태','법인형태','설립일','사업자등록번호','직원수','우편번호','회사주소','결산월']
df=pd.DataFrame(columns=cols)
df
```

```python
for i in range(98001,100001):
    
    try:
        
        driver.get(f'https://bizno.net/article/{no_bisocial[i]}')
        time.sleep(0.005)
        html = driver.page_source
        soup = BeautifulSoup(html)
        #         ths = soup.find_all("th")
        tds = soup.find_all("td")

        text=[]
        text.append(no_bisocial[i])
        text.append(soup.find_all("h4")[0].text)
        for txt in [0,1,2,3,6,7,-6,-3,-2,-1]: # 7번에서 광고배너로 한칸씩 밀림 # 예외처리를 따로 하는것보다는 사업자 등록번호 ffill로 밀어넣으면 될것같음
            text.append(tds[txt].text.strip())
        df.loc[i]=text
        time.sleep(0.005)
        
    except: continue
```
