## 대입 시작

```python
company = 매입_업종_list[0]
```

```python
매입2 = 매입1[['big_typ','da_date','mn_bungae']]
매출2 = 매출1[['big_typ','da_date','mn_bungae']]
```

```python
매입2 = 매입2[매입2['big_typ']==company]
매출2 = 매출2[매출2['big_typ']==company]
```

```python
매입2.sort_values(by='da_date', inplace=True)
매입2['da_date'] = 매입2['da_date'].astype('str')

year = [i[:6] for i in 매입2['da_date'] ]
year = list(map(int, year))
매입2['year'] = year
```

```python
# ----
매출2.sort_values(by='da_date', inplace=True)
매출2['da_date'] = 매출2['da_date'].astype('str')

year = [i[:6] for i in 매출2['da_date'] ]
year = list(map(int, year))
매출2['year'] = year
```

```python
매입2
```

```python
매출2
```

```python
year_col = [201601,201602,201603,201604,201605,201606,201607,201608,201609,201610,201611,201612]
```

```python
매입2['year_1'] = [str(i)[:4] for i in 매입2['year']]

line = Line(f"업종별 연도별 시계열 매입액 ({company})")


for i in ['2016', '2017', '2018', '2019'] :
    try:
        df_1 = 매입2[매입2['year_1']==i].sort_values('year')
        df_1 = df_1[['year','mn_bungae']]
        df_2 = df_1.groupby('year').sum()
        df_2.reset_index(inplace = True)

        year = list(df_2['year'].astype('str'))
        mount = list(df_2['mn_bungae'])
        cnt = 0
        for real, ii in zip(year, [i+str(m)[-2:] for m in year_col]) :
            if real != ii :
                year.append(ii)
                year.sort()
                mount.insert(cnt,0)        
            cnt += 1

        line.add(i, [str(m)+'월' for m in range(1,13)], list(map(lambda x: x*1/100000000,mount)), is_datazoom_show=True,datazoom_range=[0, 100] 
             , legend_text_size=15, legend_pos= 'right', mark_line=["average"]
           )
    except: continue
    
line
```
